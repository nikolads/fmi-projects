```python
%pylab inline

import sir
import matplotlib.pyplot as plt
```

```python
# Naive model of The Great Plague in London, 1666

plt.figure(figsize=(18, 4));
plt.xlabel("Time (days)");
plt.ylabel("Infectives (people)");
xticks(range(0, 501, 20))

N = 7781984
s0 = N - 1
i0 = 1
r0 = 0
infect_prob = 0.000000463
recovery_rate = 0.2

res = sir.simulate(s0, i0, r0, infect_prob, recovery_rate, 500, sim_step=0.001)

plot(res.t, res.I)    # Infected
plot(res.t, res.S)    # Susceptible
plot(res.t, res.R)    # Removed
```

```python
# A more accurate model of The Great Plague in London, 1666

plt.figure(figsize=(18, 4));
plt.xlabel("Time (days)");
plt.ylabel("Infectives (people)");
xticks(range(0, 501, 20))

N = 7781984
s0 = 0.065 * N
i0 = 20
r0 = N - s0 - i0
infect_prob = 0.000000463
recovery_rate = 0.2

res = sir.simulate(s0, i0, r0, infect_prob, recovery_rate, 500, sim_step=0.001)

plot(res.t, res.I)    # Infected
# plot(res.t, res.S)    # Susceptible
# plot(res.t, res.R)    # Removed
```
