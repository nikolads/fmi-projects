```python
%pylab inline

from sir import Sir
import matplotlib.pyplot as plt
```

```python
# Naive model of The Great Plague in London, 1666

plt.figure(figsize=(18, 4));
plt.xlabel("Days");
plt.ylabel("People");
xticks(range(0, 401, 40))

N = 7781984
s0=N-1
i0=1
r0=0

sir = Sir(s0, i0, r0, infect_prob=0.000000463, recovery_rate=0.2)
res = sir.simulate(400, sim_step=0.001)

plot(res.t, res.I, label="Infectives", linewidth=3)
# plot(res.t, res.S, label="Susceptible")
# plot(res.t, res.R, label="Removed")
plt.legend()
```

```python
# A more accurate model of The Great Plague in London, 1666

plt.figure(figsize=(18, 4));
plt.xlabel("Days");
plt.ylabel("People");
xticks(range(0, 401, 40))

N = 7781984
s0 = 0.065*N
i0 = 20
r0 = N-s0-i0

sir = Sir(s0, i0, r0, infect_prob=0.000000463, recovery_rate=0.2)
res = sir.simulate(400, sim_step=0.001)

plot(res.t, res.I, label="Infectives", linewidth=3)
# plot(res.t, res.S, label="Susceptible")
# plot(res.t, res.R, label="Removed")
plt.legend()
```
