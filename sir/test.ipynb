```python
%pylab inline

from sir import Sir
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18, 7));
plt.xlabel("Time (days)");
plt.ylabel("Infectives (people)");
xticks(range(0, 100, 1))

N = 763
sir = Sir(s0=N-1, i0=1, r0=0, infect_prob=2.18e-3, recovery_rate=202*2.18e-3)
res = sir.simulate(20, sim_step=0.001, precision=0.1)

plot(res.t, res.I, label="Infected", linewidth=2.5)
plot(res.t, res.S, label="Susceptible", linewidth=2.5)
plot(res.t, res.R, label="Removed", linewidth=2.5)
plt.legend()

res.write_to_file("test_data.csv")
```
